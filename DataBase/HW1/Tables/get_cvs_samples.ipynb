```python
import pandas as pd

file_path = '/content/customer_and_transaction.xlsx'

transaction_sheet = pd.read_excel(file_path, sheet_name='transaction')
customer_sheet = pd.read_excel(file_path, sheet_name='customer')
merged_data = pd.merge(transaction_sheet, customer_sheet, on='customer_id')

data_to_port = merged_data.sample(20)
data_to_port.head()
```

```python
customers = data_to_port[[
        'customer_id',
        'first_name',
        'last_name',
        'gender',
        'DOB',
        'job_title',
        'job_industry_category',
        'wealth_segment',
        'deceased_indicator',
        'owns_car',
        'address',
        'postcode',
        'state',
        'country',
        'property_valuation'
]]
transactions = data_to_port[[
        'transaction_id',
        'product_id',
        'customer_id',
        'transaction_date',
        'online_order',
        'order_status'
]]
products = data_to_port[[
        'product_id',
        'brand',
        'product_line',
        'product_class',
        'product_size',
        'list_price',
        'standard_cost'
]]
```

```python
customers.to_csv('customers.csv', index = False)
transactions.to_csv('transactions.csv', index = False)
products.to_csv('products.csv', index = False)
```
